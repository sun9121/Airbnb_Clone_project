```python
import pandas as pd

df = pd.read_csv("roomdetail_final_2.csv")


df.head(45)
    
```

```python
count_blank = 0
count_null = 0
null_df = df.isnull()

for i in range(len(df)):
    if df.loc[i]["room_types"] == " ":
        count_blank += 1
    if null_df["room_price"][i] == True:
        df["room_price"][i] = " "
        count_null += 1
```

```python
print(count_blank , count_null)
```
